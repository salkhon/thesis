```python
from pathlib import Path
import pandas as pd

IMGDIR = Path("/home/salkhon/Documents/thesis/data/images")

all_df = pd.DataFrame()
for lang_dir in IMGDIR.iterdir():
    if lang_dir.is_dir():
        lang = lang_dir.name
        csv_path = lang_dir/f"{lang}.csv"
        print("Reading", csv_path, csv_path.exists())
        all_df = pd.concat([all_df, pd.read_csv(csv_path)])
```

```python
print("Number of USEFUL images", len(all_df.loc[all_df["ImageStatus"] == "USEFUL"]))
print("Number of FILTERED images", len(all_df.loc[all_df["ImageStatus"] == "FILTERED"]))
```

```python
print("Number of articles having atleast one useful image")
all_df.groupby("ArticleId")["ImageStatus"].apply(lambda group: (group == "USEFUL").any()).sum()
```
